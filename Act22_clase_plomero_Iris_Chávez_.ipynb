```python
class Plomero:
    def __init__(self, licencia, herramientas, experiencia):
        # Inicialización de los atributos
        self.licencia = licencia
        self.herramientas = herramientas
        self.experiencia = experiencia

    def instalacion_de_tuberias(self):
        return f"Realizando instalación profesional con licencia {self.licencia}."

    def mantenimiento_de_tuberias(self):
        return f"Ejecutando mantenimiento preventivo usando mis {self.herramientas}."

    def reportar_estado(self):
        return f"Plomero con {self.experiencia} años de experiencia activo."

# Creación de los objetos (instancias)
fontanero1 = Plomero("LP-1020", "llave inglesa y soplete", 5)
fontanero2 = Plomero("LP-8845", "cortadora de tubos y sellador", 12)

# Uso de las funciones y visualización de datos
print("--- Datos de Fontanero 1 ---")
print(f"Licencia: {fontanero1.licencia}")
print(fontanero1.instalacion_de_tuberias())
print(fontanero1.reportar_estado())

print("\n--- Datos de Fontanero 2 ---")
print(f"Licencia: {fontanero2.licencia}")
print(fontanero2.mantenimiento_de_tuberias())
print(fontanero2.reportar_estado())
print("---------------------------------------------------")
print("Programa realizado por Iris Chavez NL = 11")
```
